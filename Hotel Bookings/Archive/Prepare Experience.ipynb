```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

## I. Get the Training/Testing Data

```python
df = pd.read_csv("hotel_bookings.csv")
```

```python
df_reduce = df.sample(10000).reset_index(drop=True)
df_train, df_test = train_test_split(df_reduce, test_size=0.2)
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
df_train["Set"] = "train"
df_test["Set"] = "test"
```

```python
# check that training and testing set have the same balance 
print("Proportion of canceled reservation in full set: {:1f}".format(df[df['is_canceled'] == 0].shape[0]/df.shape[0]*100))
print("Proportion of canceled reservation in training set: {:1f}".format(df_train[df_train['is_canceled'] == 0].shape[0]/df_train.shape[0]*100))
print("Proportion of canceled reservation in testing set: {:1f}".format(df_test[df_test['is_canceled'] == 0].shape[0]/df_test.shape[0]*100))
```

```python
df_train.to_csv("hotel_bookings_train.csv")
df_test.to_csv("hotel_bookings_test.csv")
```

# II. Prepare the Data for SF

```python
total_data = pd.concat([df_train,df_test])
```

```python
total_data.to_csv("hotel_bookings_salesforce_data.csv")
```
